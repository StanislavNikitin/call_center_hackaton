```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
# from scipy import stats

# import operator

from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.preprocessing.imputation import Imputer
# from sklearn.model_selection import train_test_split

# from scipy.stats import pearsonr as Correlation

import gc
```

```python
raw_x_tr = pd.read_csv('data/X_train.csv')
raw_y_tr = pd.read_csv('data/y_train.csv', header=None)

raw_x_t = pd.read_csv('data/X_test.csv')
```

```python
tr_single_value_columns = []
t_single_value_columns = []

for col in raw_x_tr.columns:
    if(raw_x_tr[col].unique().shape[0] == 1):
        tr_single_value_columns.append(col)
    if(raw_x_t[col].unique().shape[0] == 1):
        t_single_value_columns.append(col)
        
print(len(tr_single_value_columns), len(t_single_value_columns))
```

```python
cat_features = ['temperature1', 'temperature2', 'temperature3', 'temperature4', 
                'temperature5', 'temperature6', 'temperature7', 'temperature8', 
                'temperature9', 'temperature10', 'speed24', 'chem2', 'chem3', 
                'chem6', 'chem16', 'chem19', 'chem23', 'chem24', 'chem27', 'chem29', 
                'size5']

speed_cols = ['speed1','speed2','speed3','speed4','speed5','speed6','speed7','speed8',
             'speed9','speed10','speed11','speed12','speed13','speed14','speed15','speed16',
             'speed17','speed18','speed19','speed20',
             'speed21','speed22', 'speed23','speed24','speed25','speed26','speed27']
force_cols = ['force1','force2','force3','force4','force5','force6','force7','force8','force9','force10',
              'force11','force12', 'force13','force14','force15','force16','force17','force18','force19','force20',
             'force21','force22', 'force23','force24','force25','force26','force27','force28','force29','force30',
              'force31','force32', 'force33','force34','force35','force36','force37','force38','force39','force40',
             'force41','force42', 'force43','force44','force45','force46','force47', 'force48','force49','force50',
             'force51','force52', 'force53']
temperature_cols = ['temperature1','temperature2','temperature3','temperature4','temperature5',
                    'temperature6','temperature7','temperature8', 'temperature9','temperature10',
                    'temperature11']
power_cols = ['power1','power2','power3','power4']
chem_cols = ['chem1','chem2','chem3','chem5','chem6','chem9','chem10',
              'chem15','chem16','chem17','chem18','chem19',
             'chem21', 'chem23','chem24','chem26','chem27','chem29','chem30']
gap_cols = ['gap1','gap2','gap3','gap4','gap5','gap6','gap7','gap8','gap9','gap10',
            'gap11','gap12', 'gap13','gap14','gap15','gap16']
size_cols = ['size1','size2','size4','size5','size6','size7','size8','size9','size10',
            'size11']

col_groups = [speed_cols, force_cols, temperature_cols, power_cols, chem_cols, gap_cols, size_cols]

col_groups_dict = {speed_cols[0] : 'spd',
                  force_cols[0] : 'frc',
                  temperature_cols[0] : 'temp',
                  chem_cols[0] : 'chem',
                  gap_cols[0] : 'gap',
                  size_cols[0] : 'size',
                  power_cols[0] : 'pwr',}
```

```python
df = raw_x_tr.copy()
score_df = raw_x_t.copy()
```

```python
df_null_columns=df.columns[df.isnull().any()]
df[df_null_columns].isnull().sum()
```

```python
score_df_null_columns = score_df.columns[score_df.isnull().any()]
score_df[score_df_null_columns].isnull().sum()
```

```python
temperature_cols = ['temperature1','temperature2','temperature3','temperature4','temperature5',
                    'temperature6','temperature7','temperature8', 'temperature9','temperature10',
                    'temperature11']

df[temperature_cols] = df[temperature_cols].fillna(0.)
score_df[temperature_cols] = score_df[temperature_cols].fillna(0.)
```

```python
# fill NaNs with median values

imp = Imputer(missing_values='NaN', strategy='median', axis=0).fit( df.append(score_df) )

df_filled = pd.DataFrame(imp.transform(df), columns=df.columns)
y_filled = raw_y_tr
score_df_filled = pd.DataFrame(imp.transform(score_df), columns=score_df.columns)

# df_filled = df.fillna(-1)
# score_df_filled = score_df.fillna(-1)
```

```python
df_filled[speed_cols].std(axis=1)
```

```python
# feature generating

for group in col_groups:
    name = col_groups_dict[group[0]]
    print(name)
    
    df_filled[name+'_mean'] = df_filled[group].mean(axis=1)
    df_filled[name+'_max'] = df_filled[group].max(axis=1)
    df_filled[name+'_min'] = df_filled[group].min(axis=1)
    df_filled[name+'_std'] = df_filled[group].std(axis=1)
    
    score_df_filled[name+'_mean'] = score_df_filled[group].mean(axis=1)
    score_df_filled[name+'_max'] = score_df_filled[group].max(axis=1)
    score_df_filled[name+'_min'] = score_df_filled[group].min(axis=1)
    score_df_filled[name+'_std'] = score_df_filled[group].std(axis=1)
    
```

```python
# scale dataset across columns

scaler = StandardScaler().fit(df_filled.append(score_df_filled))

df_scaled_xgb = pd.DataFrame(scaler.transform(df_filled), columns=df_filled.columns)
score_df_scaled_xgb = pd.DataFrame(scaler.transform(score_df_filled), columns=score_df_filled.columns)

df_scaled = df_filled
score_df_scaled = score_df_filled

df_scaled.shape, df_scaled_xgb.shape
```

```python
# remove single-value columns

df_scaled_xgb = df_scaled_xgb[[col for col in df_scaled_xgb.columns if col not in tr_single_value_columns]]
score_df_scaled_xgb = score_df_scaled_xgb[[col for col in score_df_scaled_xgb.columns
                                           if col not in tr_single_value_columns]]

print(len(df_scaled_xgb.columns))
print(len(score_df_scaled_xgb.columns))
```

```python
#  5.06

ftr_num = X_xgb.shape[1]
y_scaler = RobustScaler().fit( y_xgb.reshape(-1,1) )

def baseline_model():
    model = Sequential()
    model.add(Dense(ftr_num, input_dim=ftr_num, kernel_initializer='normal', activation='relu'))
    model.add(Dense(128, kernel_initializer='random_uniform'))
    model.add(Dense(64, kernel_initializer='random_uniform'))
    model.add(Dense(8, kernel_initializer='random_uniform'))
    model.add(Dense(1, kernel_initializer='random_uniform'))
    # Compile model
    model.compile(loss='mean_absolute_error', optimizer='adam')
    return model
    
estimator = KerasRegressor(build_fn=baseline_model, epochs=100, batch_size=200, verbose=1 )
estimator.fit(X_xgb, y_scaler.transform( y_xgb.reshape(-1,1) ))

nn_prediction = y_scaler.inverse_transform( estimator.predict(score_df_scaled_xgb).reshape(-1,1) )

# saving
np.savetxt("Armijo_nn.txt", nn_prediction)
```

### Aggregate historical data

```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
raw_df = pd.read_excel("отчет.xlsx", sheet_name=4, skiprows=1)
raw_df.head(5)
```

```python
raw_df.columns
```

```python
cols = [u'Дата рейса', u'Рейс', u'Время отправления по расписанию', u'Время отправления фактическое']
```

```python
df = raw_df[cols]
df['delay'] = df[u'Время отправления фактическое'] - df[u'Время отправления по расписанию']
df.head()
```

```python
min_0 = timedelta(minutes=0)
min_30 = timedelta(minutes=30)
min_60 = timedelta(minutes=60)
min_120 = timedelta(minutes=120)

def getClass(delta):
#     if(delta <= min_0): return 0
    if(delta <= min_30): return 1
    if(delta <= min_60): return 2
    if(delta <= min_120): return 3
    return 4

df['label'] = df.delay.apply(getClass)
df = df.dropna()
```

```python
df.info()
```

```python
df.tail(10)
```

```python
df[u'Дата рейса'].max(), df[u'Дата рейса'].min()
```

```python
this_date = datetime(2018, 1,1,0,0,0)
that_date = datetime(2018, 1,1,0,0,0)

history_df = df[df[u'Дата рейса'] < this_date]
history_df.count()
```

```python
history_df.head(1)
```

```python
last_month_delay_byclass_count_df = \
    history_df[[u'Рейс', 'label', 'delay']].groupby([u'Рейс', 'label']).count().unstack().fillna(.0)

last_month_delay_byclass_count_df.columns = [col[1]  for col in last_month_delay_byclass_count_df.columns]
last_month_delay_byclass_count_df = last_month_delay_byclass_count_df.reset_index()

last_month_delay_byclass_count_df['flight'] = last_month_delay_byclass_count_df[u'Рейс']
del last_month_delay_byclass_count_df[u'Рейс']
```

```python
last_month_delay_byclass_count_df.head(10)
```

```python
last_month_delay_byclass_count_df.columns
```

```python
last_month_delay_byclass_count_df.to_csv('delay_by_class_counts_'+str(this_date)+'a.csv')
```

### Combine dataset

```python
raw_df.columns
```

```python
model_cols = [u'Дата рейса', u'Рейс', u'Тип рейса',
                u'Статус', u'А/П отправл',
               u'А/П прибыт', u'Тип ВС', u'Номер ВС',
               u'Время отправления по расписанию', 
              u'Время отправления фактическое',
              u'Время прибытия по расписанию'
               ]
model_df = raw_df[model_cols]
model_df['delay'] = model_df[u'Время отправления фактическое'] - model_df[u'Время отправления по расписанию']
```

```python
model_df['label'] = model_df.delay.apply(getClass)
model_df = model_df.dropna()

model_df.head()
```

```python
features = [u'Тип рейса',
            u'А/П отправл',
            u'А/П прибыт', u'Тип ВС', 
            u'Время отправления по расписанию',
            u'Время прибытия по расписанию']  #u'Номер ВС',
label = 'label'
features_to_code = [u'Тип рейса',u'А/П отправл', u'А/П прибыт', u'Тип ВС']  # u'Номер ВС',
date_features = [u'Время отправления по расписанию', u'Время прибытия по расписанию']
```

```python
that_date
```

```python
model_df_after_date = model_df[model_df[u'Дата рейса'] >= that_date]

dataset = model_df_after_date[features + [u'Рейс']]
labels = model_df_after_date[label]

dataset['flight'] = dataset[u'Рейс']
```

```python
# Join with history data here

dataset = pd.merge(dataset, last_month_delay_byclass_count_df,
                         how='left', on='flight')

dataset = dataset.fillna(.0)
```

```python
dataset.head(1)
```

```python
def get_parsed_data(time):
    parsed = datetime.strptime(str(time), "%Y-%m-%d %H:%M:%S") 
    return parsed.month, parsed.day, parsed.hour, parsed.weekday() #, parsed.minute, parsed.second

dataset['month_d'], dataset['day_d'], dataset['Hour_d'], dataset['weekday_d'] = \
    zip(*dataset[u'Время отправления по расписанию'].apply(get_parsed_data))
    #, dataset['M_d'], dataset['S_d'] 

dataset['month_a'], dataset['day_a'], dataset['Hour_a'], dataset['weekday_a']  = \
    zip(*dataset[u'Время прибытия по расписанию'].apply(get_parsed_data))
    #, dataset['M_a'], dataset['S_a']
```

```python
del dataset[u'Рейс']
del dataset[u'Время отправления по расписанию']
del dataset[u'Время прибытия по расписанию']
del dataset['flight']
```

```python
dataset.head()
```

```python
# encode features

def encodeColumn(col):
    values = col.unique()
    d = {}
    for i in range(values.shape[0]):
        d.update({values[i]: i})
    
    def replaceValues(value):
        return d[value]
    
    return col.apply(replaceValues)

for col in features_to_code:
    dataset[col] = encodeColumn(dataset[col])
```

```python
dataset.tail()
```

## Model

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, labels, test_size=0.2, random_state=64)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, mean_absolute_error
```

```python
parameters = {'n_estimators':[100, 200, 500], 'max_depth':[int(X_train.shape[1] / 2), X_train.shape[1], 50],
              'criterion':['gini' ,'entropy']} #, 'min_samples_split':[5, 50]}

forest = RandomForestClassifier()
clf = GridSearchCV(forest, parameters, scoring='f1_weighted')
clf.fit(X_train,  y_train)
```

```python
clf = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=42)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
prediction = clf.predict(X_test)

target_names = ['class 1', 'class 2', 'class 3', 'class 4'] # 'class 0', 
print(classification_report(y_test, prediction, target_names=target_names))
print(confusion_matrix(y_test, prediction, labels=[1,2,3,4]))
```

```python
# overfit testing

prediction_on_train = clf.predict(X_train)

target_names = ['class 1', 'class 2', 'class 3', 'class 4'] # 'class 0', 
print(classification_report(y_train, prediction_on_train, target_names=target_names))
print(confusion_matrix(y_train, prediction_on_train, labels=[1,2,3,4]))
```

```python
importances = pd.DataFrame(clf.best_estimator_.feature_importances_.T, dataset.columns.T)
importances.sort_values(by=0, ascending=False)
```

```python
def MAE(clf, x_test, y_test):
    prediction = clf.predict_proba(X_test)
    real = np.zeros(prediction.shape)
    nsamples = prediction.shape[0]
    for i in range(nsamples):
        real[i, y_test.iat[i]-1] = 1
        
    return mean_absolute_error(real, prediction)
```

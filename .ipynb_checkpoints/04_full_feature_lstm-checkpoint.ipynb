```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from datetime import date
from sklearn.metrics import mean_absolute_error
```

```python
coins = ['ethereum', 'bitcoin', 'ripple', 'litecoin', 'nem', 'stellar', 'dash']
start = date(2015,9,15)
end = date(2018, 1, 14)

# read trends

data = pd.read_csv('trends_ethereum_20150915_20180115.csv')
del data['Unnamed: 0']
data = data.rename(index=str, columns={"index": "Date", "value": "ethereum_trend"})
data.Date = data.Date.astype('datetime64[ns]')
data = data[(data.Date >= start) &  (data.Date  <= end)]

# normalize
data['ethereum_trend'] = (data.ethereum_trend - 0) / 100
data = data.reset_index()
print(len(data))
```

```python
# read coins, merge

new_df = pd.DataFrame()

for coin in coins:
    # read
    df = pd.read_csv(coin+'.csv')
    del df['Unnamed: 0']
    df.Date = df.Date.astype('datetime64[ns]')
    df = df[(df.Date >= start) &  (df.Date  <= end)]
    df = df.sort_values(['Date'], ascending=True)
    
    # normalize
    max_rate = max(df[['Open', 'Close', 'High']].max())    
    min_rate = min(df[['Open', 'Close', 'Low']].min())
    max_vol = df.Volume.max()
    min_vol = df.Volume.min()
    max_cap = df['Market Cap'].max()
    min_cap = df['Market Cap'].min()
    
    new_df[coin+'_Volatility'] = (df.High - df.Low) / df.Open # TODO: is in [0,1] ????
    new_df[coin+'_Open'] = (df.Open - min_rate) / (max_rate - min_rate + 1)
    new_df[coin+'_Close'] = (df.Close - min_rate) / (max_rate - min_rate + 1)
    new_df[coin+'_Volume'] = (df.Volume - min_vol) / (max_vol - min_vol + 1)
    new_df[coin+'_Market_Cap'] = (df['Market Cap'] - min_cap) / (max_cap - min_cap + 1)
    

new_df['ethereum_trend'] = data.ethereum_trend
new_df['Date'] = data.Date

new_df = new_df.sort_values(['Date'], ascending=False)
    
```

```python
new_df['ethereum_Close'].plot()
```

```python
# prepare X and Y

train_size = 700
test_size = len(new_df) - train_size
window_size = 20
label = 'ethereum_Close'

features = [f for f in new_df.columns if f not in ['Date']]
X = new_df[features].values
Y = new_df[label].values
```

```python
input_num = train_size - window_size
features_num = len(features)
inputs = np.zeros((input_num, window_size, features_num))
outputs = np.zeros(input_num)

print(inputs.shape)
print(outputs.shape)

index = 0
for index in range(input_num):
    inputs[index] = X[index : index + window_size]
    outputs[index] = Y[index + window_size]
print(index)
```

```python
test_inputs = np.zeros((test_size, window_size, features_num))
test_outputs = np.zeros(test_size)

j = 0
for i in range(index + 1, index + test_size + 1 ):
    test_inputs[j] = X[i : i + window_size]
    test_outputs[j] = Y[i + window_size]
    j += 1

print(test_inputs.shape)
print(test_outputs.shape)
```

## Making a model

```python
# import the relevant Keras modules
from keras.models import Sequential
from keras.layers import Activation, Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
def build_model(inputs, output_size, neurons, activ_func = "linear",
                dropout =0.25, loss="mae", optimizer="adam"):
    model = Sequential()

    model.add(LSTM(neurons, input_shape=(inputs.shape[1], inputs.shape[2])))
    model.add(Dropout(dropout))

    model.add(Dense(units=output_size))
    model.add(Activation(activ_func))

    model.compile(loss=loss, optimizer=optimizer)
    return model
```

```python
# fit the model

np.random.seed(202)
eth_model = build_model(inputs, output_size=1, neurons = 50)
eth_history = eth_model.fit(inputs, outputs, epochs=20, batch_size=5, verbose=2, shuffle=True)

f, ax = plt.subplots(1, sharex=True)
ax.plot(eth_history.epoch, eth_history.history['loss'])
plt.title('Loss')
plt.show()
```

## Testing the most recient data

```python
prediction = []

for i in range(len(test_inputs)):
    prediction.append(eth_model.predict(test_inputs)[i])
    
# real values
prediction = [float(x) * float(max_rate - min_rate) + float(min_rate) for x in prediction]
real = [float(x) * float(max_rate - min_rate) + float(min_rate) for x in test_outputs]
```

```python
plt.figure(figsize=(15,8))
pr = plt.plot(prediction, color='green', label='prediction')
re = plt.plot(real, label='real')
plt.title('Ethereum price forecasting')
plt.xlabel('days')
plt.ylabel('USD price')
plt.legend()

plt.show()

print('MAE: ', mean_absolute_error(real, prediction))
```

## Testing on random period in past

```python
def test_period(start, per_length):
    test_inputs = np.zeros((per_length, window_size, features_num))
    test_outputs = np.zeros(per_length)
    
    j = 0
    for i in range(start, per_length + start ):
        test_inputs[j] = X[i : i + window_size]
        test_outputs[j] = Y[i + window_size]
        j += 1
    
    print(test_inputs.shape)
    print(test_outputs.shape)
    
    prediction = []

    for i in range(len(test_inputs)):
        prediction.append(eth_model.predict(test_inputs)[i])
    
    # real values
    prediction = [float(x) * float(max_rate - min_rate) + float(min_rate) for x in prediction]
    real = [float(x) * float(max_rate - min_rate) + float(min_rate) for x in test_outputs]
    
    plt.figure(figsize=(10,5))
    pr = plt.plot(prediction, color='green', label='prediction')
    re = plt.plot(real, label='real')
    plt.title('Ethereum price forecasting')
    plt.xlabel('days')
    plt.ylabel('USD price')
    plt.legend()
    
    plt.show()
    print('MAE: ', mean_absolute_error(real, prediction))
```

```python
per_length = 50
start = np.random.randint(0, len(new_df) - per_length)
print(start)

test_period(start, per_length)
```

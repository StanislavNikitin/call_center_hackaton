```python
import pandas as pd
from datetime import datetime, timedelta
import gc
```

```python
# read whole raw data

df0 = pd.read_excel("in/data0.xlsx", skiprows=3, skipfooter=1)
df1 = pd.read_excel("in/data1.xlsx", skiprows=3, skipfooter=1)
df2 = pd.read_excel("in/data2.xlsx", skiprows=3, skipfooter=1)
df3 = pd.read_excel("in/data3.xlsx", skiprows=3, skipfooter=1)

raw_df = df0.append(df1).append(df2).append(df3)
raw_df = raw_df.reset_index()
del raw_df['index']
del raw_df['Unnamed: 0']

raw_df.tail(2)
```

```python
# read data with dynamic statistics 

stat_cols = ['Departure_schedule_time',
             'Jet_ID',
             'last_moth_class_count_0', 
             'last_moth_class_count_1',
       'last_moth_class_count_2', 'last_moth_class_count_3',
       'last_moth_class_count_4', 'last_year_by_class_count_0',
       'last_year_by_class_count_1', 'last_year_by_class_count_2',
       'last_year_by_class_count_3', 'last_year_by_class_count_4',
       'last_year_that_month_by_class_count_0',
       'last_year_that_month_by_class_count_1',
       'last_year_that_month_by_class_count_2',
       'last_year_that_month_by_class_count_3',
       'last_year_that_month_by_class_count_4']
             
statistic_df = pd.read_csv('out/at_most_all_only_for_Stanislav.csv')[stat_cols]
statistic_df['Departure_schedule_time'] = pd.to_datetime(statistic_df['Departure_schedule_time'])
statistic_df.tail(2)
```

```python
# select meaningful columns

model_cols = [u'Дата рейса', u'Рейс', u'Тип рейса',
            u'Статус', u'А/П отправл',
            u'А/П прибыт', u'Тип ВС', u'Номер ВС',
            u'Время отправления по расписанию', 
            u'Время отправления фактическое',
            u'Время прибытия по расписанию'
            ]

model_df = raw_df[model_cols]
model_df['delay'] = model_df[u'Время отправления фактическое'] - model_df[u'Время отправления по расписанию']
```

```python
min_0 = timedelta(minutes=0)
min_15 = timedelta(minutes=15) 
min_60 = timedelta(minutes=60)
min_120 = timedelta(minutes=120)

def getClass(delta):
    if(delta <= min_0): return 0
    if(delta <= min_15): return 1
    if(delta <= min_60): return 2
    if(delta <= min_120): return 3
    return 4

model_df['label'] = model_df.delay.apply(getClass)
model_df = model_df.dropna()
model_df.head(2)
```

```python
# take dataset since 2017

start_date = datetime(2017, 1,1,0,0,0)

model_df = model_df[model_df[u'Дата рейса'] > start_date]
statistic_df = statistic_df[statistic_df['Departure_schedule_time'] > start_date]
print(model_df.count())
print(statistic_df.count())
```

```python
# merge model_df with statistic_df

df = pd.merge(model_df, statistic_df,  how='left',
              left_on=[u'Номер ВС', u'Время отправления по расписанию'],
              right_on=['Jet_ID', 'Departure_schedule_time'])
df = df.dropna()
```

```python
df.columns
```

```python
# extract date and time features

def get_parsed_data(time):
    parsed = datetime.strptime(str(time), "%Y-%m-%d %H:%M:%S") 
    return parsed.month, parsed.day, parsed.hour, parsed.weekday()

df['month_d'], df['day_d'], df['Hour_d'], df['weekday_d'] = \
    zip(*df[u'Время отправления по расписанию'].apply(get_parsed_data))

df['month_a'], df['day_a'], df['Hour_a'], df['weekday_a']  = \
    zip(*df[u'Время прибытия по расписанию'].apply(get_parsed_data))
```

```python
# fuck russian letters!

df['Flight'] = df[u'Рейс']
df['Flight_type'] = df[u'Тип рейса']
df['airport_dep'] = df[u'А/П отправл']
df['airport_arr'] = df[u'А/П прибыт']
df['Jet_type'] = df[u'Тип ВС']
```

```python
# delete excessive and duplicate columns

del df[u'Дата рейса']
del df[u'Статус']
del df[u'Номер ВС']
del df['delay']
del df[u'Время отправления по расписанию']
del df[u'Время отправления фактическое']
del df[u'Время прибытия по расписанию']
del df[u'Departure_schedule_time']

del df[u'Рейс']
del df[u'Тип рейса']
del df[u'А/П отправл']
del df[u'А/П прибыт']
del df[u'Тип ВС']
```

```python
# encode categorical features

features_to_code = ['Flight',
                    'Flight_type', 
                    'airport_dep', 
                    'airport_arr', 
                    'Jet_type', 
                    'Jet_ID']


def encodeColumn(col):
    values = col.unique()
    d = {}
    for i in range(values.shape[0]):
        d.update({values[i]: i})
    
    def replaceValues(value):
        return d[value]
    
    return col.apply(replaceValues)
```

```python
# simple encoding for trees
for col in features_to_code:
    df[col] = encodeColumn(df[col])

## hard one hot
# df = pd.get_dummies(df, columns=features_to_code)
```

```python
df.tail(2)
```

```python
# polynomial columns

columns = ['Flight',
           'Flight_type',
           'airport_dep',
           'airport_arr',
           'Jet_type',
           'Jet_ID',
           'month_d',
           'day_d',
           'Hour_d',
           'weekday_d',
           'month_a',
           'day_a',
           'Hour_a',
           'weekday_a']


processed = []
polynomial_features = []

for col1 in columns:
    for col2 in columns:
        if col2 not in processed and col1 != col2:
            name = str(col1+'---'+col2)
            polynomial_features.append(name)
            df[name] = df[col1].astype(str) + '___' + df[col2].astype(str)
            processed.append(col1)


more_features_to_code = features_to_code + polynomial_features
```

```python
## simple encoding of polynomial features

for col in polynomial_features:
    df[col] = encodeColumn(df[col])
```

```python
# hard one hot
for col in more_features_to_code:
    df = pd.get_dummies(df, columns=[col])
```

```python
features = [col for col in df.columns] # if col not in ['label'] ]

dataset = df[features]
labels = df['label']
```

## Model

```python
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import xgboost

from sklearn.metrics import classification_report, confusion_matrix, mean_absolute_error
```

```python
# force garbage collector
gc.collect()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, labels, test_size=0.2, random_state=64)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
def balance(data):
    factor = 2

    counts = []
    classes = data.label.unique()

    for cl in classes:
        counts.append(data[data.label == cl].count()[0])

    minimal = min(counts)
    print(minimal)

    dataset1 = pd.DataFrame()
    for cl in classes:
        d = data[data.label == cl].sample(int(factor*minimal), replace=True)
        dataset1 = dataset1.append(d)
    
    dataset1 = dataset1.sample(frac=1).reset_index(drop=True)

    return dataset1, dataset1['label']

X_train, y_train = balance(X_train)
X_test, y_test = balance(X_test)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
del X_train['label']
del X_test['label']
```

```python
# Grid Search over Random Forests

parameters = {'n_estimators':[100], 'max_depth':[6, int(X_train.shape[1] / 5)],
              'criterion':['gini']} #, 'min_samples_split':[5, 50]}

forest = RandomForestClassifier()
clf = GridSearchCV(forest, parameters, scoring='f1_weighted')
clf.fit(X_train,  y_train)

clf.best_params_
```

```python
# Random Forest

clf = RandomForestClassifier(n_estimators=100,
                             max_depth=int(8), # X_train.shape[1]*0.2
                             bootstrap=True,
                             max_features=0.5,
                             oob_score=True,
                             random_state=42)
clf.fit(X_train, y_train)
```

```python
# XGBoost

clf = xgboost.XGBClassifier(objective='multi:softprob',
                            eval_metric="rmse")
clf.fit(X_train, y_train)
```

```python
# prediction

prediction = clf.predict(X_test)

target_names = ['class 0', 'class 1', 'class 2', 'class 3', 'class 4'] # 'class 0', 
print(classification_report(y_test, prediction, target_names=target_names))
print(confusion_matrix(y_test, prediction, labels=[0,1,2,3,4]))
print

# overfit testing

prediction_on_train = clf.predict(X_train)

target_names = ['class 0', 'class 1', 'class 2', 'class 3', 'class 4'] # 'class 0', 
print(classification_report(y_train, prediction_on_train, target_names=target_names))
print(confusion_matrix(y_train, prediction_on_train, labels=[0, 1,2,3,4]))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from datetime import date
from sklearn.metrics import mean_absolute_error, mean_squared_error
import talib as ta
```

```python
# collecting data

coins = ['ethereum', 'bitcoin', 'ripple'] #, 'litecoin', 'nem', 'stellar', 'dash']
start = date(2015,9,15)
end = date(2018,3,1)

# read trends

# data = pd.read_csv('daily/trends_ethereum_20150915_20180301.csv')
# del data['Unnamed: 0']
# data = data.rename(index=str, columns={"index": "Date", "value": "ethereum_trend"})
# data.Date = data.Date.astype('datetime64[ns]')
# data = data[(data.Date >= start) &  (data.Date  < end)]


label = 'ripple_Close'
target_coin = 'ripple'
max_target_rate = 0
min_target_rate = 0

# normalize
# data['ethereum_trend'] = (data.ethereum_trend - 0) / 100
# data = data.reset_index()

# read coins, merge

new_df = pd.DataFrame()

for coin in coins:
    # read
    df = pd.read_csv('daily/'+coin+'.csv')
    del df['Unnamed: 0']
    df.Date = df.Date.astype('datetime64[ns]')
    df = df[(df.Date >= start) &  (df.Date  < end)]
    df = df.sort_values(['Date'], ascending=True)
    
    stoch, rsi = ta.STOCHRSI(df.Close.values, timeperiod=14)
    
    # normalize
    if(coin == target_coin):
        max_target_rate = max(df[['Open', 'Close', 'High']].max()) 
        min_target_rate = min(df[['Open', 'Close', 'Low']].min())
    max_rate = max(df[['Open', 'Close', 'High']].max())    
    min_rate = min(df[['Open', 'Close', 'Low']].min())
    max_vol = df.Volume.max()
    min_vol = df.Volume.min()
    max_cap = df['Market Cap'].max()
    min_cap = df['Market Cap'].min()
    
    new_df[coin+'_Volatility'] = (df.High - df.Low) / df.Open # TODO: is in [0,1] or in [-1,1] ?
    new_df[coin+'_Open'] = (df.Open - min_rate) / (max_rate - min_rate + 1)
    new_df[coin+'_Close'] = (df.Close - min_rate) / (max_rate - min_rate + 1)
    new_df[coin+'_Volume'] = (df.Volume - min_vol) / (max_vol - min_vol + 1)
#     new_df[coin+'_Market_Cap'] = (df['Market Cap'] - min_cap) / (max_cap - min_cap + 1)
    new_df[coin+'_Stoch'] = stoch.tolist()
    new_df[coin+'_RSI'] = rsi.tolist()
    new_df['Date'] = df['Date']

new_df = new_df.reset_index() 
    
# new_df['ethereum_trend'] = data.ethereum_trend
# new_df['Date'] = data.Date

new_df = new_df.sort_values(['Date'], ascending=True)
del new_df['index']

new_df = new_df.dropna()
for coin in coins:
    new_df[coin+'_Stoch'] = (new_df[coin+'_Stoch'] - new_df[coin+'_Stoch'].mean()) / (new_df[coin+'_Stoch'].max() - new_df[coin+'_Stoch'].min())
    new_df[coin+'_RSI'] = (new_df[coin+'_RSI'] - new_df[coin+'_RSI'].mean()) / (new_df[coin+'_RSI'].max() - new_df[coin+'_RSI'].min())

new_df = new_df.reset_index() 
del new_df['index']
```

```python
new_df.head()
```

```python
# prepare X and Y

new_df = new_df[new_df.Date > date(2017, 1, 1)]

train_size = 300 # 800
test_size = len(new_df) - train_size

window_size = 15

features = [f for f in new_df.columns if f not in ['Date']]
X = new_df[features].values
Y = new_df[label].values

input_num = train_size - window_size
features_num = len(features)
inputs = np.zeros((input_num, window_size, features_num))
outputs = np.zeros(input_num)

print(inputs.shape)
print(outputs.shape)

index = 0
for index in range(input_num):
    inputs[index] = X[index : index + window_size]
    outputs[index] = Y[index + window_size]
print(index)

test_inputs = np.zeros((test_size, window_size, features_num))
test_outputs = np.zeros(test_size)

j = 0
for i in range(index + 1, index + test_size + 1 ):
    test_inputs[j] = X[i : i + window_size]
    test_outputs[j] = Y[i + window_size]
    j += 1

print(test_inputs.shape)
print(test_outputs.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

## LSTM

```python
# import the relevant Keras modules
from keras.models import Sequential
from keras.layers import Activation, Dense
from keras.layers import LSTM, GRU
from keras.layers import Dropout
```

```python
def build_model(inputs, output_size, neurons, activ_func = "linear", # linear
                dropout =0.5, loss='mse', optimizer="adam"): # 'mse'
    model = Sequential()

#     model.add(LSTM(neurons, return_sequences=True, input_shape=(inputs.shape[1], inputs.shape[2])))
#     model.add(LSTM(neurons))
    model.add(GRU(neurons, return_sequences=True, input_shape=(inputs.shape[1], inputs.shape[2])))
    model.add(GRU(neurons))

    model.add(Dropout(dropout))

    model.add(Dense(units=output_size))
    model.add(Activation(activ_func))

    model.compile(loss=loss, optimizer=optimizer)
    return model


# fit the model

np.random.seed(2)
eth_model = build_model(inputs, output_size=1, neurons = 150)
eth_history = eth_model.fit(inputs, outputs, epochs=15, batch_size=20, verbose=1, shuffle=True) 
```

```python
def custom_accuracy(y_true, y_pred):
    diff_true = np.sign(y_true[1:] - y_true[0:-1])
    diff_pred = np.sign(y_pred[1:] - y_pred[0:-1])
    return len([i for i in range(len(diff_true)) if diff_true[i] == diff_pred[i]]) / len(diff_true)
```

```python
prediction = []

for i in range(len(test_inputs)):
    prediction.append(eth_model.predict(test_inputs)[i])
    
# real values
prediction = [float(x) * float(max_target_rate - min_target_rate +1) + float(min_target_rate) for x in prediction]
real = [float(x) * float(max_target_rate - min_target_rate +1) + float(min_target_rate) for x in test_outputs]

plt.figure(figsize=(15,8))
pr = plt.plot(prediction, color='red', label='prediction')
re = plt.plot(real, label='real')
plt.title('Ripple price forecasting')
plt.xlabel('days')
plt.ylabel('USD price')
plt.legend()

plt.show()
print('accuracy: ', custom_accuracy(np.array(real), np.array(prediction)))
print('MAE: ', mean_absolute_error(real, prediction))
print('RMSE: ', np.sqrt(mean_squared_error(real, prediction)))
```
